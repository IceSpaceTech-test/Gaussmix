# Gaussian Mixture Demo

Synthetic fitting and model-selection walkthrough.

```python
import numpy as np
from gaussian_mixture import GaussianMixture

rng = np.random.default_rng(42)
X = np.vstack([
    rng.normal(loc=(-2, 0), scale=0.45, size=(150, 2)),
    rng.normal(loc=(2, 0), scale=0.55, size=(150, 2)),
])
model = GaussianMixture(n_components=2, covariance_type='full', random_state=42)
model.fit(X)
model.score(X), model.predict(X[:10])
```

```python
from gaussian_mixture import information_criteria

scores = []
for k in range(1, 5):
    candidate = GaussianMixture(n_components=k, random_state=42).fit(X)
    scores.append((k, information_criteria(candidate, X, criterion='bic')))
scores
```
